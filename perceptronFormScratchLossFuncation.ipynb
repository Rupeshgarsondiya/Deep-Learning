```python
# import required library
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# load data
X,y = make_classification(n_classes=2,class_sep=5,n_samples=100,n_features=2,hypercube=False,random_state=30,n_informative=1,n_redundant=0,n_clusters_per_class=1)
print(X)
print(y)
```

```python
# plot the data
plt.scatter(X[:,0],X[:,1],c=y)
```

```python

def perceptron(X,y):
    w1=w2=b=1
    lr = 0.55  # here learning rate is 0.55
    for j in range(50):
        for i in range(X.shape[0]):

            z = (w1*X[i][0]+w2*X[i][1]+b)*y[i]
            if z<0:
                w1 = w1 + lr * X[i][0] * y[i]
                w2 = w2 + lr * X[i][1] * y[i]
                b = b + lr * y[i]
                print(w1,w2,b)

    return w1,w2,b

w1,w2,b = perceptron(X,y)
m = -(w1/w2)
c = -(b/w2)
print('slope : ',m)
print('intersect : ',c)
import numpy as np
x_input = np.linspace(-4,4,100)
y_input = m*x_input+c

plt.plot(x_input,y_input,linewidth=3,color = 'red')
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
p.fit(X,y)
print('slope : ',p.coef_)
print('intersect : ',p.intercept_)
```
