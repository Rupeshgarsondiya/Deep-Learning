# Vanishing Gardient Problem In Deep Neural Network

```
# import required library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import make_moons
```

```
# load dataset
X, y = make_moons(n_samples=100, random_state=1)

print(X.shape)
print(y.shape)

# plot the data
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='rainbow')
plt.show()
```

# Vanishing Gradient Problem

**Defination of vanishing gradient problem :**

- In machine learning, the vanishing gradient problem is encountered when training neural networks with gradient-based learning methods and backpropagation. In such methods, during each training iteration, each neural network weight receives an update proportional to the partial derivative of the loss function with respect to the current weight. The problem is that as the network depth or sequence length increases, the gradient magnitude typically is expected to decrease (or grow uncontrollably), slowing the training process. In the worst case, this may completely stop the neural network from further learning.
                                                               
                                                               
                                                                                                          - According to Wikipidia

- In simple terms, when training a neural network using backpropagation, we calculate gradients.
- For deep neural networks, the gradient of the input layer is the result of multiplying multiple gradients. Generally, the gradient values lie between zero and one. When these gradients are multiplied repeatedly, the resulting value approaches zero. This is why there is almost no change in the weights and biases.
- For this reason, there is no significant change in the loss function, and the neural network training stops. This phenomenon is called the vanishing gradient problem.

# Vanishing gradient problem

```
# Design the deep neural network

model = Sequential()
model.add(Dense(10, input_dim=2, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add((Dense(10, activation='sigmoid')))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
model.summary()


old_weight = model.get_weights()[0][0]
# compile model

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# fit model

history1 = model.fit(X, y, epochs=100,validation_split=0.2)

new_weight = model.get_weights()[0][0]

print(f'Old weight : {old_weight}')
print(f'New weight : {new_weight}')

# plot the graph

plt.plot(history1.history['loss'], label='Training loss')
plt.plot(history1.history['val_loss'], label='Validation loss')
plt.legend()
plt.show()

plt.plot(old_weight, label='old weight')
plt.plot(new_weight, label='new weight')
plt.legend()
plt.show()

# Here our loss is constant not any change in our loss
```

**According to the loss graph, we can clearly see that our neural network is facing the vanishing gradient problem. The training loss remains constant, with little to no change in the weights and biases.**

# Solution

# 1 . Use shallow neural network

- This solution does not work for every problem, but for small datasets, we can use a shallow neural network instead of a deep neural network.
- However, even in shallow neural networks, if we use sigmoid or tanh as activation functions, it can still cause the vanishing gradient problem.

```
# Design shallow neural network
model1 = Sequential()
model1.add(Dense(5, input_dim=2, activation='sigmoid'))
model1.add(Dense(3, activation='sigmoid'))
model1.add(Dense(2, activation='sigmoid'))
model1.add(Dense(1, activation='sigmoid'))
model1.summary()

old_weight1 = model1.get_weights()[0][0]

model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

history1 = model1.fit(X, y, epochs=100, validation_split=0.2)

new_weight1 = model1.get_weights()[0][0]

print(f'Old weight : {old_weight1}')
print(f'New weight : {new_weight1}')

# plot the graph of loss and weight
plt.plot(history1.history['loss'], label='Training loss')
plt.plot(history1.history['val_loss'], label='Validation loss')
plt.legend()
plt.show()

plt.plot(old_weight1, label='old weight')
plt.plot(new_weight1, label='new weight')
plt.legend()
plt.show()
```

**Here, we use a shallow neural network, but there is still not much change in the loss .**

# 2. ReLU as activation

- This technique work well
- We train deep neural network use activation function as ReLU     (Rectified Linear Unit)

```
# Design Deep neural network

model2 = Sequential()
model2.add(Dense(10, input_dim=2, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))
model2.summary()

old_weight2 = model2.get_weights()[0][0]

model2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

history2 = model2.fit(X, y, epochs=100, validation_split=0.2)

new_weight2 = model2.get_weights()[0][0]

print(f'Old weight : {old_weight2}')
print(f'New weight : {new_weight2}')

# plot loss and weight
plt.plot(history2.history['loss'], label='Training loss')
plt.plot(history2.history['val_loss'], label='Validation loss')
plt.legend()
plt.show()

plt.plot(old_weight2, label='old weight')
plt.plot(new_weight2, label='new weight')
plt.legend()
plt.show()
```

**Use ReLU as activation loss is extremly dicrese .**
